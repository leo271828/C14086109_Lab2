```python
import pygame
import time

WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
ENY_WIDTH = 50
ENY_HEIGHT = 50
HP_WIDTH = 40
HP_HEIGHT = 40
FPS = 30

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()

# load image (background, enemy, buttons)
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load('images/enemy.png'), (ENY_WIDTH, ENY_HEIGHT))
pause_image = pygame.transform.scale(pygame.image.load('images/pause.png'), (BTN_WIDTH, BTN_HEIGHT))
continue_image = pygame.transform.scale(pygame.image.load('images/continue.png'), (BTN_WIDTH, BTN_HEIGHT))
sound_image = pygame.transform.scale(pygame.image.load('images/sound.png'), (BTN_WIDTH, BTN_HEIGHT))
muse_image = pygame.transform.scale(pygame.image.load('images/muse.png'), (BTN_WIDTH, BTN_HEIGHT))

hp_image = pygame.transform.scale(pygame.image.load('images/hp.png'), (HP_WIDTH, HP_HEIGHT))
hp_gray_image = pygame.transform.scale(pygame.image.load('images/hp_gray.png'), (HP_WIDTH, HP_HEIGHT))

# set the clock
clock = pygame.time.Clock()

# set the title
pygame.display.set_caption('My first game')

class Game:
    def __init__(self):
        # window
        self.window = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
        
        # time
#         self.time 是在算 while 總共跑幾次，while每執行一次就將 self.time + 1
#         self.time_str 則是要輸出在畫面上的字串
#         self.font 則是指定 self.time_str 等等在畫面上的字體為 Arial
            
        self.time = 0
        self.time_str = '00:00'
        self.font = pygame.font.SysFont("arial", 24)
        
        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):
        # game loop
        run = True
        while run:
            clock.tick(FPS)
            # event loop
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False

            # draw background
            self.window.blit(background_image, (0, 0))

            # draw enemy and health bar
            self.window.blit(enemy_image, (50, 250))
            pygame.draw.rect(self.window, (255, 0, 0), [50, 230, ENY_WIDTH, 8])

            # draw menu (and buttons)
            pygame.draw.rect(self.window, (0, 0, 0), [0, 0, WIN_WIDTH, BTN_HEIGHT])
            self.window.blit(pause_image, (WIN_WIDTH - BTN_WIDTH, 0))
            self.window.blit(continue_image, (WIN_WIDTH - 2*BTN_WIDTH, 0))
            self.window.blit(sound_image, (WIN_WIDTH - 3*BTN_WIDTH, 0))
            self.window.blit(muse_image, (WIN_WIDTH - 4*BTN_WIDTH, 0))
            self.window.blit(sound_image, (WIN_WIDTH - 3*BTN_WIDTH, 0))
            
#             init 為第一個 hp的 x位置
#             x_loc 就是其他 hp的 x位置
#             y_loc 為第一個 hp的 y位置
#             算法是一排有5格圖案 所以每五個圖案 y 方向就要向下加一行
            for idx in range(self.max_hp):
                init = WIN_WIDTH/2 - 2*HP_WIDTH
                x_loc = init + (idx % 5)*HP_WIDTH
                y_loc = (idx // 5)*HP_HEIGHT
                
#                 看你有多少血量 剩餘血量就是用有顏色的圖案 被扣掉的血量就是沒有顏色的圖案
                if idx < self.hp :
                    self.window.blit(hp_image, (x_loc, y_loc))
                else :
                    self.window.blit(hp_gray_image, (x_loc, y_loc))

#             draw time
#             因為 self.time 在算跑過幾次 while 所以每次要 +1
#             FPS = 30 則代表一秒內程式會執行 while 迴圈 30次，大概啦 :D
#             因此如果 self.time = 30  代表 經過 1 秒
#                 如果 self.time = 60  代表 經過 2 秒
#                      ...
#                     self.time = 30*n 代表 經過 n 秒
# 
#             得出 只要 self.time 為 FPS 的 n 倍時，代表現在過了 n 秒鐘
            self.time += 1
            
#             利用剛剛得出的結論來計算遊戲的時間
#             其中 zfill(2) 是用於補 0 ，直到數字變成 2 位數
#                 ex:  1 -> 01
#                      3 -> 03
#                     12 -> 12
            if self.time%FPS == 0:
                minutes = str(self.time // (FPS*60)).zfill(2)
                second = str(self.time // FPS).zfill(2)
                self.time_str = minutes + ":" + second
            
#             製造一個拿來放時間的長方形
            white = 255, 255, 255
            pygame.draw.rect(self.window, (0, 0, 0), [0, WIN_HEIGHT-40, 100, 40])
            text_surface = self.font.render(self.time_str, True, white)
            self.window.blit( text_surface, (25, WIN_HEIGHT-35) ) 
            
            pygame.display.update()
        pygame.quit()
        
if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()


```
